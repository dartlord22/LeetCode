Given the root of a binary tree, invert the tree, and return its root.

Example:



Input: root = [4,2,7,1,3,6,9],
Output: [4,7,2,9,6,3,1]

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def reverseTree(self, root):
        if root is None:
            return None
        # Swap the left and right children
        root.left, root.right = root.right, root.left
        # Recur for left and right subtrees
        self.reverseTree(root.left)
        self.reverseTree(root.right)
        return root

def createTreeFromList(values):
    if not values:
        return None
    
    # Create the root of the tree
    root = TreeNode(values[0])
    nodes = [root]
    index = 1
    
    # Use a queue to keep track of nodes to attach children to
    while index < len(values):
        current = nodes.pop(0)
        
        if index < len(values) and values[index] is not None:
            current.left = TreeNode(values[index])
            nodes.append(current.left)
        index += 1
        
        if index < len(values) and values[index] is not None:
            current.right = TreeNode(values[index])
            nodes.append(current.right)
        index += 1
        
    return root

def levelOrderTraversal(root):
    if not root:
        return []
    
    result = []
    queue = [root]
    
    while queue:
        current = queue.pop(0)
        result.append(current.val)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    
    return result

# Input
nums = [4, 2, 7, 1, 3, 6, 9]
root = createTreeFromList(nums)

# Reverse the tree
tree = TreeNode()
tree.reverseTree(root)

# Output
output = levelOrderTraversal(root)
print(output)  # Output: [4, 7, 2, 9, 6, 3, 1]

```
